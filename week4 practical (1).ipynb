## <U>1) Consider auto-mpg dataset and perform univariate</U>

## A) Define Univariate
### Univariate data refers to a dataset where each observation is associated with only one variable. This means it focuses on measuring or observing a single characteristic or attribute for each individual in the dataset.
## Example : 


## <U>Using Auto-Mpg Dataset</U>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/Adim/Downloads/bhoomika/auto-mpg.csv')
df
```

## <u>Using Univariate Ploting Histogram</u>:-

```python
plt.hist(df['cylinders'])
plt.title('counting cylinders')
plt.xlabel('cylinders')
plt.ylabel('cylinders counts')
```

## <u>Using Univariate Ploting Line Graph</u> :-

```python
df['model year'].value_counts().sort_index().plot(kind='line')
plt.xlabel('Model Year')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Model Year')
plt.show()
```

 ## <u>Using Univariate Ploting Boxplot</u> :-

```python
plt.boxplot(df['weight'])
plt.title('MPG Distribution by weight')
plt.ylabel('weight')
plt.show()
```

 ## <u>Using Univariate Ploting Bar Graph</u> :-

```python
df['cylinders'].value_counts().plot(kind='bar')
```

## <u>Using Univariate Ploting Pie Chart </u>:-

```python
df['origin'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('count origins')
```

## <U>2) Consider Titanic Dataset And Perform Bivariate</U>
## <u>B) Define Bivariate</u>

### Bivariate means the analysis of two variables. Using bivariate analysis we can find how well the variables are correlated. 

### Example:


## <u>Using Titanic Dataset</u> :-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/Adim/Downloads/bhoomika/titanic.csv')
df
```

## <u>Using Uivariate Ploting Line Graph</u> :-


```python
a=df.groupby(['survived','sex']).size().unstack()
a.plot(kind='line',marker='*')
```

## <u>Using Univariate Ploting Bar Plot</u> :-

```python
a=df.groupby(['sex','survived']).size().unstack()
a.plot(kind='bar')
```

## <u>Using Bivariate Ploting Scatter Plot</u> :-


```python
plt.scatter(df['age'], df['fare'])
plt.xlabel('Age')
plt.ylabel('fare')
plt.title('Survival vs Age')
plt.show()
```
